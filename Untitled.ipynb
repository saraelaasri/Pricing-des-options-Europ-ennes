```python
# Initialise parameters
S = dt['Close'][0]    # initial stock price
K =  dt['Close'].mean()     # strike price
F=  dt['Close'].shape
T=F[0]/252 # time to maturity in years #T = days_to_maturity / 252

r = 0.06      # annual risk-free rate
n= dt['Close'].shape
N=n[0]# number of time steps
dt['Log returns'] = np.log(dt['Close']/dt['Close'].shift())
sigma=dt['Log returns'].std()*252**.5 # Annualised stock price volatility
```

```python


def Cox_Ross_Rubinstein_Tree (S,K,T,r,sigma,N, Option_type):
    
    
    u=math.exp(sigma*math.sqrt(T/N));
    d=math.exp(-sigma*math.sqrt(T/N));
    pu=((math.exp(r*T/N))-d)/(u-d);
    pd=1-pu;
    disc=math.exp(-r*T/N);

    St = [0] * (N+1)
    C = [0] * (N+1)
    
    St[0]=S*d**N;
    
    for j in range(1, N+1): 
        St[j] = St[j-1] * u/d;
    
    for j in range(1, N+1):
        if Option_type == 'P':
            C[j] = max(K-St[j],0);
        elif Option_type == 'C':
            C[j] = max(St[j]-K,0);
    
    for i in range(N, 0, -1):
        for j in range(0, i):
            C[j] = disc*(pu*C[j+1]+pd*C[j]);
            
    return C[0]

```

```python
def Jarrow_Rudd_Tree (S,K,T,r,sigma,N, Option_type):

    
        
    u=math.exp((r-(sigma**2/2))*T/N+sigma*math.sqrt(T/N));
    d=math.exp((r-(sigma**2/2))*T/N-sigma*math.sqrt(T/N));
    pu=0.5;
    pd=1-pu;
    disc=math.exp(-r*T/N);

    St = [0] * (N+1)
    C = [0] * (N+1)
    
    St[0]=S*d**N;
    
    for j in range(1, N+1): 
        St[j] = St[j-1] * u/d;
    
    for j in range(1, N+1):
        if Option_type == 'P':
            C[j] = max(K-St[j],0);
        elif Option_type == 'C':
            C[j] = max(St[j]-K,0);
    
    for i in range(N, 0, -1):
        for j in range(0, i):
            C[j] = disc*(pu*C[j+1]+pd*C[j]);
            
    return C[0]
```

```python
def d11(S, X, T, r, sigma):
    return (np.log(S/X) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))

#Fonctions complmentaires   
def d21(d1, T, sigma):   
    return d1 - sigma * np.sqrt(T)

   
def bs_delta(S, X, T, r, sigma, option_type):
    if option_type == 'call':
        return norm.cdf(d11(S, X, T, r, sigma))
    elif option_type == 'put':
        return norm.cdf(-d11(S, X, T, r, sigma))
    else:
        raise ValueError("Option type is either 'call' or 'put'.")

# BLACK SCHOLES  
def black_scholes(S, X, T, r, sigma, option_type):
   
    d_one = d11(S, X, T, r, sigma)
    d_two = d21(d_one, T, sigma)
    if option_type == 'call':
        return S * norm.cdf(d_one) - np.exp(-r * T) * X * norm.cdf(d_two)
    elif option_type == 'put':
        return -(S * norm.cdf(-d_one) - np.exp(-r * T) * X * norm.cdf(-d_two))
    else:
       
        raise ValueError("Option type is either 'call' or 'put'.")
```

```python
# Modle trinomiale
def TrinomialTreeOption (AmeEurFlag, CallPutFlag, S, X, Time, r, sigma, n):
    z=0
    if CallPutFlag == "c": 
        z  =  +1 
    if CallPutFlag == "p": 
        z  =  -1 
    dt  =  Time/n;
    OptionValue  =  [0] * (2*n+5)
    
    u  =  math.exp(+sigma * math.sqrt(2*dt));
    d  =  math.exp(-sigma * math.sqrt(2*dt));
    #  Les probabilits up and down:
    pu  =  ((math.exp(r * dt/2) - math.exp( -sigma * math.sqrt(dt/2))) / 
               (math.exp(sigma * math.sqrt(dt/2)) - math.exp(-sigma * math.sqrt(dt/2)))) ** 2;
    pd  =  (( math.exp(sigma * math.sqrt(dt/2)) - math.exp( r* dt/2)) / 
            (math.exp(sigma * math.sqrt(dt/2)) - math.exp(-sigma * math.sqrt(dt/2)))) ** 2;
   #  La probilit pour rester dans la meme position:
    pm  =  1 - pu - pd
    Df  =  math.exp(-r*dt)   
    for i in range(2*n+2):
        OptionValue[i+1]  =  max(0, z*(S*u**max(i-n, 0) * 
            d**max(n*2-n-i, 0) - X))
    for j in reversed(range(n+1)):
        for k in range(2*j):
               #Type de l'option
            if AmeEurFlag == "e":
                OptionValue[k+1]  =  (pu * OptionValue[k+3] +pm * OptionValue[k+2] +pd * OptionValue[k+1]) * Df  
              
            elif AmeEurFlag == "a":
                OptionValue[k+1]  =  max((z*(S*u^max(k-j, 0) *d ^ max(j*2-j-k, 0) - X)), (pu * OptionValue[k+3] 
                                                                                 + pm * OptionValue[k+2] +pd * OptionValue[k+1]) * Df)  
                        
                        
                         
                        
    TrinomialTree  =  OptionValue[1] 
        
       
    return TrinomialTree


    
```
